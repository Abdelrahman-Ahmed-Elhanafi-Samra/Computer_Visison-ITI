```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from datetime import datetime
import math
```

# <font color = orange>Function used ðŸ“š

```python
def printDate(img):
    
    now = datetime.now()
    
    
    font = cv2.FONT_HERSHEY_SIMPLEX
    
    cv2.putText(img, f'Date : {now.year}-{now.month}-{now.day}  [ { now.strftime("%A") } ]',
                (0,30), font, 0.80, (245, 244, 18), 2, cv2.LINE_AA)
    cv2.imshow('Clock', img)

    ##############################################

def printTime( img):
    
        
    second = datetime.now().strftime("%S")
    minute = datetime.now().strftime("%M")
    hour = datetime.now().strftime("%I")

    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img, f'Time : {hour}:{minute}:{second}', (0,65), font, 0.80, (245, 244, 18), 2, cv2.LINE_AA)
    cv2.imshow('Clock', img)
    
    ##############################################
    
def hourPoint(radius, img, name_window):
    
    for degree in range(12): 
        x = int((250 + 150 * math.cos(math.radians(radius))))
        y = int((250 + 150 * math.sin(math.radians(radius))))
                
        cv2.circle(img, (x, y) ,3, (0, 255, 255) , 2)
        cv2.imshow(name_window, img)
        radius += 30
        
    ##############################################

def minutePoint(radius, img, name_window):
    
    for degree in range(60):
        x = int((250 + 150*math.cos(math.radians(radius))))
        y = int((250 + 150*math.sin(math.radians(radius))))
        
        cv2.circle(img, (x,y) ,2, (255, 255, 255) , 1)
        cv2.imshow(name_window, img)
        radius += 6
        
    ##############################################
    
def secondHand(red, green, blue):
    
    now = datetime.now()
    second = datetime.now().strftime("%S")
    angle_second = ((int(second) * 6) - 90)
    
    x = int((250 + 140*math.cos(math.radians(angle_second))))
    y = int((250 + 140*math.sin(math.radians(angle_second))))
    
    cv2.line(img, (250, 250), (x, y), (blue, green, red))
    
    ##############################################
    
def minuteHand(red, green, blue):
    
    now = datetime.now()
    minute = datetime.now().strftime("%M")
    angle_minute = (((int(minute) * 6) - 90))
    
    x = int((250 + 120*math.cos(math.radians(angle_minute))))
    y = int((250 + 120*math.sin(math.radians(angle_minute)))) 
    
    cv2.line(img, (250, 250), (x, y), (blue, green, red), 2)
    
    ##############################################
    
def hourHand(red, green, blue):
    
    now = datetime.now()
    hour = datetime.now().strftime("%I")
    
    angle_hour = (((int(hour) * 30) - 90))
    
    x = int((250 + 100*math.cos(math.radians(angle_hour))))
    y = int((250 + 100*math.sin(math.radians(angle_hour))))
    
    cv2.line(img, (250, 250), (x, y), (blue, green, red), 3)
    
    
```

```python
cv2.namedWindow("Clock")
radius = 0


    
while True:
    
    
    img = np.zeros((500, 500, 3), np.uint8)
    
    # primary circuit
    
    cv2.circle(img, (250, 250), 150, (0,0,255) ,1)

    
    # loop for Hour point
    
    hourPoint(radius, img, 'Clock')
    
    
    # loop for Minute point    
        
    minutePoint(radius, img, 'Clock')   


    
    secondHand(255, 255, 0)
    
    minuteHand(104, 245, 10)
    
    hourHand(1, 99, 96)
    

    
    printDate(img)
    printTime(img)
    
    
    # Circle of center of clock
    
    cv2.circle(img, (250, 250) ,5, (86, 80, 78) , -1)
    cv2.imshow("Clock",img)
    
    
    
    k = cv2.waitKey(1000)
    if k == ord('q'):
        break
    
cv2.destroyAllWindows()
```
