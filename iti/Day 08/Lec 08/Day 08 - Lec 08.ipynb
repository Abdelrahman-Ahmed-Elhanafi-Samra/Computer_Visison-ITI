```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = np.zeros((250,250))
plt.imshow(img, cmap='gray')
```

```python
img[50:151, 50] = 255
plt.imshow(img, cmap='gray')
```

```python
np.fill_diagonal(img, 255)
plt.imshow(img, cmap='gray')
```

```python
img2 = np.fliplr(img)
np.fill_diagonal(img2, 255)
img2 = np.fliplr(img2)
plt.imshow(img2, cmap='gray');
```

```python
image = np.zeros((255,255,3))
plt.imshow(image, cmap='gray')
```

```python
image[50:151,50,0] = 255
image[50:151,100,1] = 255
image[50:151,150,2] = 255
plt.imshow(image)
```

```python
image2 = np.zeros((256,256,3), np.uint8)

for i in range(256):
    image2[50,i,0] =  i
    image2[100,i,1] = i
    image2[150,i,2] = i
plt.imshow(image2)
```

```python
image3 = np.zeros((256,256,3), np.uint8)

image3[:,:,0][50] = np.arange(0, 256)
image3[:,:,1][50] = np.arange(255, -1, -1)
image3[:,:,1][100] = np.arange(255, -1, -1)
image3[:,:,0][150] = np.arange(0, 256)
image3[:,:,1][150] = np.arange(0, 128, 0.5)

plt.imshow(image3)
```
