```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Q1 load the following image and make red rectangle around face and write below rectangle ‚ÄúCat‚Äù with yellow color.
_____________________________________

```python
cat = cv2.imread('original.png')

width =  int(cat.shape[0] * 0.5)
height = int(cat.shape[1] * 0.3)

resized_cat = cv2.resize(cat, (width, height), interpolation = cv2.INTER_AREA)


cv2.rectangle(resized_cat, (290,45),(600,330),(0,0,255),3)

cv2.putText(resized_cat, "Cat", (370, 400), cv2.FONT_HERSHEY_PLAIN, 5, (0, 255, 255), 4, cv2.LINE_AA)

cv2.imshow('Image', resized_cat)

plt.imshow(resized_cat[:, :, ::-1])

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Q2 load the following image and apply sharpen filter. 

## ‚Äúwrite your observation‚Äù
____________________________________

```python
blur_cat = cv2.imread('blurred.png')

kernel = np.array([[-1,-1,-1],
                   [-1, 9,-1],
                   [-1,-1,-1]])


sharp = cv2.filter2D(blur_cat, -1, kernel, borderType = cv2.BORDER_ISOLATED)

f, (axis0, axis1) = plt.subplots(1, 2, figsize=(30,15))

axis0.set_title("Orignal")
axis0.imshow(blur_cat[:, :, ::-1])

axis1.set_title("Image with Filter")
axis1.imshow(sharp[:, :, ::-1])


'''
Through my observation of the two images, I did not find a big difference between the original image 
and the image after adding the filter to it, but there is almost a clarification of the image
a little after adding the filter, but it is not clear enough.
'''
```
